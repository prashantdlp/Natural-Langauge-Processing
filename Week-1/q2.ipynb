```python
import numpy as np
```

```python
arr = np.random.uniform(0,1,20) * 10
print("arr: ",arr)
```

```python
rounded = np.round(arr,2)
print("rounded",rounded)
```

```python
max = rounded.max()
min = rounded.min()
median = np.median(rounded)
print("max:",max,"\nmin:",min,"\nmedian:",median)
```

```python
def numpy_alternate_sort(array) :
    n = len(array)
    forward  = sorted(array)
    backward = sorted(array, key = lambda x : -x)
    for i in range(n):
        if i % 2  == 0 : 
            array[i] = forward[i//2]
        else :
            array[i] = backward[i//2]
    return array            
  
print(numpy_alternate_sort(rounded))
```
