```python
import numpy as np
```

```python
arr = np.random.randint(1,50,(5,4)) # should have been 51 instead of 50
print("arr: ",arr)
```

```python
print(arr[0,-1],arr[1,-2],arr[2,-3],arr[3,-4],end="\n")

anti_diag = arr[np.arange(min(arr.shape)), arr.shape[1] - 1 - np.arange(min(arr.shape))]
print("\nAnti-diagonal elements:",anti_diag)
```

```python

max = np.max(arr,axis = 1)
print("max :",max)
```

```python
mean = np.mean(arr)
print("mean:",mean )

new_arr = arr[arr <=  mean]
print("new_arr :",new_arr)
```

```python
def numpy_boundary_traversal(matrix):
    # no error handeling done here
    list = []
    list.append(matrix[0,:])
    list.append(matrix[1:,-1])
    list.append(matrix[-1,-2:0:-1])
    list.append(matrix[-1:0:-1,0])
    return list 

print("traversal :",numpy_boundary_traversal(arr))
```
